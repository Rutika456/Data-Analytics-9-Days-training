```python
import matplotlib.pyplot as plt
week_names=["week-1","week-2","week-3","week-4"]
sale_amount=[1000,2000,3000,20000]

study_hours=[1,2,3,6]
marks=[20,32,13,56]

mark_1=[1,2,3,4,54,45,32,65,78,90,23,43,53,76]

plt.figure(figsize=(12,8))
plt.subplot(2,2,1)
#--------------------------subplot-1--------------------------

plt.plot(week_names,sale_amount,color="red",marker="D",mfc="k",mec="y",ls="-.",lw=10)
plt.xlabel("weeks")
plt.ylabel("sales")
plt.grid()

#--------------------------subplot-2--------------------------

plt.subplot(2,2,2)
plt.bar(week_names,sale_amount,color="red",hatch="/\*",edgecolor="m")
plt.xlabel("weeks")
plt.ylabel("slaes")
plt.grid()

#--------------------------subplot-3--------------------------

plt.subplot(2,2,3)
plt.scatter(study_hours,marks,c=marks,cmap="rainbow",s=300)
plt.xlabel("study hours")
plt.ylabel("marks")
plt.grid()

#--------------------------subplot-4---------------------
plt.subplot(2,2,4)
plt.hist(mark_1,bins=10,color="red",rwidth=0.5)
plt.xlabel("marks")
plt.ylabel("count")
plt.grid()

plt.show()
```
