```python
import matplotlib.pyplot as plt

Teams=["rcb","kkr","gi","csk","mi"]
winning_counts=[40,60,10,20,15]
plt.plot(Teams,winning_counts,color="c",ls="--",lw=7,marker="H",mfc="y",mec="k",ms=15,mew=4)
plt.title("point table")
plt.xlabel("Teams")
plt.ylabel("winning_count")
plt.grid(color="y",ls=":",lw=2)
plt.savefig("pointtable.png")
plt.show()
```

```python
import matplotlib.pyplot as plt

semister=["sem-1","sem-2","sem-3","sem-4","sem-5","sem-6"]
backlogs=[40,60,10,20,15,30]
plt.plot(semister,backlogs,color="c",ls="-",lw=7,marker="H",mfc="y",mec="k",ms=15,mew=4)
plt.title("reslt analysis")
plt.xlabel("semister")
plt.ylabel("backlogs")
plt.xticks(rotation=90)
plt.grid(color="r",ls="-.",lw=2)
plt.savefig("result.png")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
data={"Days": ["mon","tue","wed","thu","fri","sat","sun"],
    "week_1":[120,340,254,643,123,345,234],
    "week_2":[130,350,154,443,523,645,334],
    "week_3":[140,540,254,543,223,545,434]
    }
df=pd.DataFrame(data)
plt.plot(df["Days"],df["week_1"],color="k",label="week-1",lw="10")
plt.plot(df["Days"],df["week_2"],color="r",label="week-2",lw="10")
plt.plot(df["Days"],df["week_3"],color="g",label="week-3",lw="10")
plt.legend(loc="lower right")
plt.grid()
plt.xticks(rotation=90)
plt.xlabel("Days")
plt.ylabel("Steps")
plt.title("Steps Analysis")
plt.show()
```

```python
import matplotlib.pyplot as plt

Teams=["rcb","kkr","gi","csk","mi"]
winning_counts=[60,40,10,20,15]
colors=["red","green","blue","yellow","orange"]
plt.bar(Teams,winning_counts,color=colors,edgecolor="black",linewidth=3,width=0.5,hatch="*",)
plt.xlabel("Teams")
plt.ylabel("winning_count")
plt.title("point table")
plt.show()
```

```python
import matplotlib.pyplot as plt

Teams=["rcb","kkr","gi","csk","mi"]
winning_counts=[40,60,10,20,15]
colors=["red","green","blue","yellow","orange"]
plt.barh(Teams,winning_counts,color=colors,edgecolor="black",hatch="xo",)
plt.xlabel("Teams")
plt.ylabel("winning_count")
plt.title("point table")
plt.show()
```

```python
import matplotlib.pyplot as plt

studyhours=[1,3,4,5]
marks=[23,45,67,54]
plt.figure(figsize=(5,5))
plt.scatter(studyhours,marks,color=["red","blue","green","yellow"],marker="D",s=150,edgecolor="black",linewidth=3)
plt.grid(color="red")
plt.xlabel("studyhours")
plt.ylabel("marks")
plt.title("studyhours vs marks")
plt.show()
```
