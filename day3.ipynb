##copy and view

```python
arr=np.array(np.arange(1,5))
print(arr)
```

```python
arr_1=arr.copy()
print(arr_1)
```

```python
arr_1[0]=100
print(arr_1)
print(arr)
```

```python
print(np.shares_memory(arr,arr_1))
```

```python
arr=np.array(np.arange(1,5))
print(arr)
arr_2=arr.view()
print(arr_2)
print(np.shares_memory(arr,arr_2))
```
