```python
rows = 5
for i in range(1, rows + 1):
    asterisks_string = " ".join(["*"] * i)
    total_width = 2 * rows - 1
    leading_spaces = total_width - len(asterisks_string)
    print(" " * leading_spaces + asterisks_string)
```
